```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
from sklearn.svm import SVC

style.use('fivethirtyeight')

# Create mesh grids
def make_meshgrid(x, y, h=.02):
    x_min, x_max = x.min()-1, x.max()+1
    y_min, y_max = y.min()-1, y.max()+1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy

# Plot the contours
def plot_contours(ax, clf, xx, yy, **params):
    z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    z = z.reshape(xx.shape)
    out = ax.contourf(xx, yy, z, **params)
    return out

subset_size = 500

# Features and target
features = df[['Magnitude(ergs)', 'Latitude(deg)']][:subset_size].values
classes = df['Magnitude_type'][:subset_size].values

# Create SVMs
svm1 = SVC(kernel='rbf')
svm2 = SVC(kernel='rbf')
svm3 = SVC(kernel='rbf')

# Train SVMs
svm1.fit(features, (classes == 'ML').astype(int))
svm2.fit(features, (classes == 'Mx').astype(int))
svm3.fit(features, (classes == 'Md').astype(int))

# Meshgrid for plotting
fig, ax = plt.subplots()
x0, x1 = features[:, 0], features[:, 1]
xx, yy = make_meshgrid(x0, x1)

# Plot decision boundaries
plot_contours(ax, svm1, xx, yy, cmap=plt.cm.Reds, alpha=0.2)
plot_contours(ax, svm2, xx, yy, cmap=plt.cm.Blues, alpha=0.2)
plot_contours(ax, svm3, xx, yy, cmap=plt.cm.Greens, alpha=0.2)

# Color setup
color = ['y', 'b', 'g', 'k', 'm']

# Scatter data points
for i in range(subset_size):
    if classes[i] == 'ML':
        plt.scatter(features[i][0], features[i][1], s=20, c=color[0], label='ML' if i == 0 else "")
    elif classes[i] == 'Mx':
        plt.scatter(features[i][0], features[i][1], s=20, c=color[1], label='Mx' if i == 0 else "")
    elif classes[i] == 'Md':
        plt.scatter(features[i][0], features[i][1], s=20, c=color[2], label='Md' if i == 0 else "")
    else:
        plt.scatter(features[i][0], features[i][1], s=20, c=color[3], label='Other' if i == 0 else "")

plt.xlabel('Magnitude (ergs)')
plt.ylabel('Latitude (deg)')
plt.title('SVM Decision Boundaries for Earthquake Magnitude Types')
plt.legend()
plt.show()
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
import matplotlib.pyplot as plt
```

```python
#read csv
df=pd.read_csv("D:/downloads/Earthquake_Data.csv",sep=r'\s+')
new_column_names = ["Date(YYYY/MM/DD)",  "Time(UTC)", "Latitude(deg)", "Longitude(deg)", "Depth(km)", "Magnitude", 
                    "Magnitude_Category", "No_of_Stations", "Gap", "Close", "RMS", "SRC", "EventID"]
df.columns=new_column_names
#convert magnitude column to categorical data
df['Magnitude_Category']=pd.cut(df['Magnitude'],bins=[0,5,6,7,np.inf],labels=['Minor','Moderate','Strong','Major'])

#Encode Magnitude Category
le=LabelEncoder()
df['Magnitude_Category_Encoded']=le.fit_transform(df['Magnitude_Category'])

#Normalize latitude and longitude values
scaler=MinMaxScaler()
df[['Latitude(deg)','Longitude(deg)']]=scaler.fit_transform(df[['Latitude(deg)','Longitude(deg)']])

#Select features
x=df[['Latitude(deg)','Longitude(deg)','No_of_Stations']]
y=df['Magnitude_Category_Encoded']

#split the data into training and testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)

#Train the Gaussian Naive Bayes model on the training data
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
y_pred=gnb.predict(x_test)
y_pred
```

```python
# Example: new earthquake data (normalized Latitude, Longitude, No_of_Stations)
new_data = [[0.45, 0.32, 12]]  # substitute with realistic values

# Predict category
new_pred_encoded = gnb.predict(new_data)
new_pred_label = le.inverse_transform(new_pred_encoded)

print("Predicted Magnitude Category:", new_pred_label[0])
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))

# Handle case where some classes might be missing in y_test
unique_labels = np.unique(y_test)
target_names = le.inverse_transform(unique_labels)

print("\nClassification Report:")
print(classification_report(y_test, y_pred, labels=unique_labels, target_names=target_names))

# Confusion matrix
conf_mat = confusion_matrix(y_test, y_pred, labels=unique_labels)
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues', 
            xticklabels=target_names, yticklabels=target_names)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Naive Bayes - Confusion Matrix")
plt.show()
```

```python
#Scatter Plot for Actual Magnitude Categories
plt.figure(figsize=(10,6))
sns.scatterplot(x=x_test['Longitude(deg)'], y=x_test['Latitude(deg)'], hue=le.inverse_transform(y_test), palette='viridis', s=30)
plt.title("Actual Earthquake Magnitude Categories")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(title="Actual Category")
plt.grid(True)
plt.show()
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

df=pd.read_csv("D:/downloads/Earthquake_Data.csv",sep=r'\s+')

df['Datetime']=pd.to_datetime(df['Date(YYYY/MM/DD)']+' '+df['Time'])
df.drop(['Date(YYYY/MM/DD)','Time'],axis=1,inplace=True)


Column_name=["Latitude(deg)", "Longitude(deg)", "Depth(km)", "Magnitude(ergs)", 
                    "Magnitude_type", "No_of_Stations", "Azimuthal_Gap", "Closest_Station", "RMS", "SRC", "EventID","DateTime"]
df.columns=Column_name

df.drop_duplicates(inplace=True)


x=df[['Latitude(deg)','Longitude(deg)','Depth(km)','No_of_Stations']]
y=df['Magnitude(ergs)']
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)


# Initialize a random forest regressor with 100 trees
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the regressor to the training data
rf.fit(x_train, y_train)
     
```

# Predict the testing data and evaluate it

```python
# Predict the target variable on the test data
y_pred = rf.predict(x_test)

# Evaluate the performance of the model using mean squared error and R^2 score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

scores['mse'].append(mse)
scores['R^2'].append(r2)

print('Mean Squared Error: ', mse)
print('R^2 Score: ', r2)
```

```python

# Plot the predicted and actual values
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Magnitude')
plt.ylabel('Predicted Magnitude')
plt.title('Random Forest Regression Results')
plt.show()
```

```python
importances = rf.feature_importances_
features = ['Latitude', 'Longitude', 'Depth', 'No. of Stations']
plt.bar(features, importances)
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Feature Importance Plot')
plt.show()
```
